```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
import sys
all_data = '../prices/prices.csv'
all_data = pd.read_csv(all_data,header = None)
data = all_data.values
import matplotlib
matplotlib.rcParams.update({
                            # Use the Computer modern font
                            'text.usetex': False,
                            'font.family': 'stixgeneral',
                            'mathtext.fontset': 'stix',
                            })
```

```python
x = data[0][0:168]
t = np.arange(x.shape[0])
fig,ax = plt.subplots(figsize = (4,2.5))
ax.plot(t,x,'k')
ax.grid(True)
ax.set_ylabel('price ($ / kWh)')
ax.set_xlabel('time (h)')

strFile = "../tex_writing/fig_1.png"
if os.path.isfile(strFile):
    os.remove(strFile)
plt.savefig(strFile,format = 'png',dpi = 600,bbox_inches = 'tight')
plt.show()
```

```python
x = data[0][0:168]
x_new = data[0][168:168 + 168]
t = np.arange(x.shape[0])
t_new = np.arange(x_new.shape[0]) + x.shape[0]
fig,ax = plt.subplots(figsize = (4,2.5))
ax.plot(t,x,'k',label = 'Training')
ax.plot(t_new,x,'r',label = 'Forecast')
ax.grid(True)
ax.legend()
ax.set_ylabel('price ($ / kWh)')
ax.set_xlabel('time (h)')

strFile = "../tex_writing/fig_2.png"
if os.path.isfile(strFile):
    os.remove(strFile)
plt.savefig(strFile,format = 'png',dpi = 600,bbox_inches = 'tight')
plt.show()
```
